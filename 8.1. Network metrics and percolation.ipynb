Theoretical estimation for the number of nodes in LCC: 
$$n*(1-q) = n*(1-e^{<k>(q-1)})$$

```python
import numpy as np 
import matplotlib.pyplot as plt

def calculate_q(k):
    '''
    Use a numerical root finder to determine q from the equation
    q = exp(k*(q-1)).
    '''
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]

LCC = []
k_average_degree = np.linspace(1,10,100)

n= 50
for k in k_average_degree: 
    q = calculate_q(k)
    LCC.append(n*(1-q))
    
plt.figure(figsize=(12,8))
plt.plot(k_average_degree, LCC)
plt.xlabel('Average Degree')
plt.ylabel('LCC Size 50 nodes')
plt.show()
```
